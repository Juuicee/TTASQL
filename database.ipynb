```python
# Function to generate random part ID
def generate_part_id():
    return random.randint(10000000, 99999999)

def get_weight(car_part):
    try:
        return next(w for part, w in weights_kg if part == car_part)
    except StopIteration:
        # If car part is not found in weights_kg list, return 0
        return 0

# Initialize empty lists to store data
part_data = []

# Define the number of rows for the dataset
num_rows = 200

# Generate random data for each row
for _ in range(num_rows):
    car_type = random.choice(car_types)
    brand = random.choice(brand_types)
    car_part = random.choice(car_parts)
    weight = get_weight(car_part)
    part_id = generate_part_id()
    part_cost = round(random.uniform(3000, 10000), 2)
    num_parts = random.randint(1, 10)
    part_data.append([part_id, car_type, brand, car_part, weight, num_parts, part_cost])

# Create DataFrame from the generated data
columns = ['Part ID', 'Car Type', 'Brand', 'Car Part', 'Weight (kg)', 'Number of Parts', 'Part Cost']
df = pd.DataFrame(part_data, columns=columns)
```
